# Bar
- Bar Graph = Comparing Categories

| Condition / Situation                | Use Bar Graph? | Explanation (Simple)                |
| ------------------------------------ | -------------- | ----------------------------------- |
| Need to compare different categories | ✅ Yes          | Example: Maths vs Science marks     |
| Data is discrete (fixed groups)      | ✅ Yes          | Example: cities, products, classes  |
| Want to compare values visually      | ✅ Yes          | Easy to see which is bigger/smaller |
| Need simple and clear presentation   | ✅ Yes          | Best for reports and presentations  |
| Want to show trend over time         | ❌ No           | Line graph is better                |
| Want to show percentage/proportion   | ❌ No           | Pie chart is better                 |


# A bar graph is used to:

- Compare values across categories

- Quickly understand differences in data

- Present data in a clear and visual way

# Color + Hatch Combo

```python
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
visitors = [300, 450, 500, 380, 620, 700, 650]

colors = ['#4CAF50','#2196F3','#FF9800','#9C27B0','#F44336','#009688','#795548']
hatches = ['/', '\\', '|', '-', '+', 'x', 'o']

bars = plt.bar(days, visitors, color=colors, edgecolor='black')

for bar, hatch in zip(bars, hatches):
    bar.set_hatch(hatch)

plt.title("Weekly Visitors")
plt.xlabel("Days")
plt.ylabel("Visitors")

plt.show()
```

# Legend with Hatch

```python
import matplotlib.patches as mpatches
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
visitors = [300, 450, 500, 380, 620, 700, 650]

colors = ['#4CAF50','#2196F3','#FF9800','#9C27B0','#F44336','#009688','#795548']
hatches = ['/', '\\', '|', '-', '+', 'x', 'o']

bars = plt.bar(days, visitors, color=colors, edgecolor='black')

legend_patches = []

for bar, hatch, color, day in zip(bars, hatches, colors, days):
    bar.set_hatch(hatch)
    patch = mpatches.Patch(facecolor=color, hatch=hatch, label=day)
    legend_patches.append(patch)

plt.legend(handles=legend_patches, title="Days")

plt.title("Weekly Visitors")
plt.xlabel("Days")
plt.ylabel("Visitors")

plt.show()
```

# Clustered Bar Chart

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

day = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

y = [300, 450, 500, 380, 620, 700, 650]
y1 = np.random.randint(1, 800, 7)

x = np.arange(len(day))   # positions
width = 0.35              # bar width

plt.figure(figsize=(8,5))

# First bar
ax = plt.bar(x - width/2, y,
             width,
             color=sns.color_palette('magma'),
             hatch='//',
             label="Traffic")

# Second bar
ax1 = plt.bar(x + width/2, y1,
              width,
              color=sns.color_palette('viridis'),
              hatch='..',
              label="Random")

# Labels
plt.title("WEEKLY WEBSITE TRAFFIC", color='g')
plt.xlabel("DAY", color='g')
plt.ylabel("VISITORS", color='b')

# Ticks
plt.xticks(x, day, color='b')
plt.yticks(color='r')

# Value labels
plt.bar_label(ax, padding=3)
plt.bar_label(ax1, padding=3)

plt.grid(axis='y', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

day = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

y1 = [300, 450, 500, 380, 620, 700, 650]
y2 = np.random.randint(100, 800, 7)
y3 = np.random.randint(50, 700, 7)

x = np.arange(len(day))
width = 0.25  

plt.figure(figsize=(9,5))

# Bars
b1 = plt.bar(x - width, y1, width,
             color=sns.color_palette('magma'),
             hatch='//',
             label="Dataset 1")

b2 = plt.bar(x, y2, width,
             color=sns.color_palette('viridis'),
             hatch='..',
             label="Dataset 2")

b3 = plt.bar(x + width, y3, width,
             color=sns.color_palette('coolwarm'),
             hatch='xx',
             label="Dataset 3")

# Labels
plt.title("WEEKLY TRAFFIC COMPARISON", color='g')
plt.xlabel("DAY", color='g')
plt.ylabel("VISITORS", color='b')

# Ticks
plt.xticks(x, day, color='b')
plt.yticks(color='r')

# Value labels
plt.bar_label(b1, padding=2)
plt.bar_label(b2, padding=2)
plt.bar_label(b3, padding=2)

plt.grid(axis='y', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

# Stacking

# Pie Graph (Pie Chart)

What is a Pie Chart?

A pie chart is a circular graph divided into slices, where:

each slice represents a part of the whole

the full circle = 100%

```python
# Pie chart shows how a whole is divided into parts
```

# When to Use a Pie Chart (IMPORTANT)

- Use a pie chart only when these conditions are satisfied:
- Data represents parts of a whole(market share, budget distribution, survey results)
- Limited number of categories(Too many slices = confusing)
- You want to show proportion, not exact comparison(Focus = percentage share)

| Feature    | Pie Chart    | Bar Graph    |
| ---------- | ------------ | ------------ |
| Purpose    | Proportion   | Comparison   |
| Total      | Must be 100% | Not required |
| Categories | Few          | Many         |


```python
# Use pie chart when you want to show percentage distribution of a whole
```

# plt.pie(x, labels=None, colors=None, autopct=None, startangle=0)
- x → values
- labels → names
- colors → slice colors 
- autopct → percentage show

```python
labels = ["Mobile", "Laptop", "Tablet"]
sizes = [60, 25, 15]
colors = ['gold', 'skyblue', 'lightgreen']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)

plt.title("Device Usage Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["Mobile", "Laptop", "Tablet", "Desktop", "Smart TV"]
sizes = [45, 20, 10, 15, 10]

colors = ['gold', 'skyblue', 'lightgreen', 'lightcoral', 'violet']

plt.figure(figsize=(6,6))

plt.pie(sizes,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops=dict(width=0.4),  #  donut effect
        shadow=True)

plt.title("Device Usage Donut Chart")

plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.pie(sizes,
        labels=labels,
        autopct='%1.1f%%',
        colors=colors,
        startangle=90,
        wedgeprops=dict(width=0.4))


plt.text(0, 0, "Usage", ha='center', va='center', fontsize=12)

plt.title("Device Usage Donut Chart")

plt.show()
```

| Parameter       | Type       | Default | Kaam (Explanation)                      | Example           |
| --------------- | ---------- | ------- | --------------------------------------- | ----------------- |
| `x`             | ArrayLike  | —       | Data values (main input)                | `[60, 25, 15]`    |
| `explode`       | ArrayLike  | None    | Slice ko bahar nikalta hai              | `[0.1, 0, 0]`     |
| `labels`        | list       | None    | Slice ke names                          | `["A","B","C"]`   |
| `colors`        | list       | None    | Slice colors set karta hai              | `['red','blue']`  |
| `autopct`       | str / func | None    | Percentage display karta hai            | `'%1.1f%%'`       |
| `pctdistance`   | float      | 0.6     | % text kitna center ke paas hoga        | `0.5`             |
| `shadow`        | bool       | False   | Shadow effect add karta hai             | `True`            |
| `labeldistance` | float      | 1.1     | Labels kitne door honge                 | `1.2`             |
| `startangle`    | float      | 0       | Pie ko rotate karta hai                 | `90`              |
| `radius`        | float      | 1       | Pie ka size control                     | `1.2`             |
| `counterclock`  | bool       | True    | Direction (anti-clockwise ya clockwise) | `False`           |
| `wedgeprops`    | dict       | None    | Slice style (width, edgecolor etc.)     | `dict(width=0.4)` |
| `textprops`     | dict       | None    | Text style (font size, color)           | `{'fontsize':10}` |
| `center`        | tuple      | (0,0)   | Pie ka position shift karta hai         | `(0,1)`           |
| `frame`         | bool       | False   | Frame (box) show karta hai              | `True`            |
| `rotatelabels`  | bool       | False   | Labels rotate karta hai                 | `True`            |
| `normalize`     | bool       | True    | Data ko 100% normalize karta hai        | `False`           |
| `hatch`         | str / list | None    | Pattern fill karta hai                  | `'//'`            |
| `data`          | any        | None    | External dataset reference              | `data=df`         |
