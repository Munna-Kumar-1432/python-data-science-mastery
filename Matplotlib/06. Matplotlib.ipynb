```python
# A Box Plot (also called a Box-and-Whisker Plot) is a statistical graph used to show the distribution of data.

# It summarizes data using:

# Minimum value

# Maximum value

# Median (middle value)

# Quartiles (Q1, Q2, Q3)
```

```python
# Box → from Q1 to Q3 (Interquartile Range)

# Line inside box → Median

# Whiskers → Minimum and Maximum

# Dots → Outliers
```

# How to Calculate Q1, Q2, Q3
- Always sort the data in ascending order
- Find Median (Q2) - (Median = middle value)
- Find Q1 (Lower Quartile) - (Q1 = median of lower half (left side)
- Find Q3 (Upper Quartile) - (Q3 = median of upper half (right side)
- 

```python
# If number of values is odd → exclude median when finding Q1 & Q3

# If number of values is even → split equally into two halves

# Q2 = Median

# Q1 = Median of lower half

# Q3 = Median of upper half
```

| Parameter        | What it does                   | Example                                                |
| ---------------- | ------------------------------ | ------------------------------------------------------ |
| `x`              | Data to plot                   | `plt.boxplot([1,2,3,4])`                               |

| `usermedians`    | Custom median value            | `plt.boxplot(data, usermedians=[5])`                   |
| `conf_intervals` | Custom CI range                | `plt.boxplot(data, conf_intervals=[[2,6]])`            |
| `meanline`       | Show mean as line              | `plt.boxplot(data, meanline=True, showmeans=True)`     |
| `showmeans`      | Show mean point                | `plt.boxplot(data, showmeans=True)`                    |
| `showcaps`       | Show whisker caps              | `plt.boxplot(data, showcaps=False)`                    |
| `showbox`        | Hide/show box                  | `plt.boxplot(data, showbox=False)`                     |
| `showfliers`     | Show outliers                  | `plt.boxplot(data, showfliers=False)`                  |
| `boxprops`       | Style of box                   | `plt.boxplot(data, boxprops=dict(color='red'))`        |
| `tick_labels`    | Labels on axis                 | `plt.boxplot([d1,d2], tick_labels=['A','B'])`          |
| `flierprops`     | Style outliers                 | `plt.boxplot(data, flierprops=dict(marker='o'))`       |
| `medianprops`    | Style median line              | `plt.boxplot(data, medianprops=dict(color='blue'))`    |
| `meanprops`      | Style mean                     | `plt.boxplot(data, meanprops=dict(marker='D'))`        |
| `capprops`       | Style caps                     | `plt.boxplot(data, capprops=dict(color='green'))`      |
| `whiskerprops`   | Style whiskers                 | `plt.boxplot(data, whiskerprops=dict(linestyle='--'))` |
| `manage_ticks`   | Auto ticks control             | `plt.boxplot(data, manage_ticks=False)`                |

```python
data = [10, 20, 30, 40, 100]

plt.boxplot(
    data,
    notch=True,
    patch_artist=True,
    showmeans=True,
    boxprops=dict(facecolor='lightblue'),
    medianprops=dict(color='red'),
    meanprops=dict(marker='D')
)

plt.title("Custom Box Plot")
plt.show()
```

# A Violin Plot is a data visualization that shows the distribution of numerical data. It combines:

- Box plot (median, quartiles)
- Density plot (shape of distribution)

```python
# Use it when you want to:

# See data distribution

# Compare multiple groups

# Understand skewness / peaks
```

```python
import matplotlib.pyplot as plt

# Data
class_A = [50, 55, 60, 65, 70]
class_B = [40, 50, 60, 80, 90]

data = [class_A, class_B]

# Create violin plot
plt.violinplot(data)

# Add labels
plt.xticks([1, 2], ['Class A', 'Class B'])
plt.ylabel("Marks")
plt.title("Violin Plot Example")

plt.show()
```

```python
# plt.violinplot(data)
# → Takes a list of datasets

# plt.xticks([1,2], [...])
# → Labels for each violin

# Default positions start from 1, not 0
```

```python
class_A = [50, 55, 60, 65, 70]
class_B = [40, 50, 60, 80, 90]

data = [class_A, class_B]

plt.violinplot(
    data,
    showmeans=True,
    showmedians=True,
    showextrema=True
)

plt.xticks([1, 2], ['Class A', 'Class B'])
plt.ylabel("Marks")
plt.title("Customized Violin Plot")

plt.show()
```

```python
# showmeans=True → shows mean line

# showmedians=True → shows median

# showextrema=True → shows min/max
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# assume df already created

fig, axes = plt.subplots(3, 3, figsize=(18, 15))

# ------------------ 1. Histogram ------------------
axes[0,0].hist(df["Runs"], color='skyblue', edgecolor='black')
axes[0,0].set_title("Runs Distribution (Histogram)")
axes[0,0].set_xlabel("Runs")
axes[0,0].set_ylabel("Frequency")

# ------------------ 2. Box Plot ------------------
axes[0,1].boxplot(df["Runs"])
axes[0,1].set_title("Runs Consistency (Box Plot)")
axes[0,1].set_ylabel("Runs")

# ------------------ 3. Violin Plot ------------------
parts = axes[0,2].violinplot(df["Runs"], showmeans=True, showmedians=True)
for pc in parts['bodies']:
    pc.set_facecolor('lightgreen')
axes[0,2].set_title("Runs Distribution Shape (Violin)")
axes[0,2].set_ylabel("Runs")

# ------------------ 4. Scatter ------------------
axes[1,0].scatter(df["Strike_Rate"], df["Runs"])
axes[1,0].set_title("Runs vs Strike Rate")
axes[1,0].set_xlabel("Strike Rate")
axes[1,0].set_ylabel("Runs")

# ------------------ 5. Line Plot ------------------
axes[1,1].plot(df["Match_ID"], df["Runs"], marker='o')
axes[1,1].set_title("Performance Over Matches")
axes[1,1].set_xlabel("Match ID")
axes[1,1].set_ylabel("Runs")

# ------------------ 6. Bar Chart ------------------
df.groupby("Match_Format")["Runs"].mean().plot(
    kind='bar', ax=axes[1,2]
)
axes[1,2].set_title("Average Runs by Format")
axes[1,2].set_xlabel("Format")
axes[1,2].set_ylabel("Average Runs")

# ------------------ 7. Fours vs Sixes ------------------
axes[2,0].scatter(df["Fours"], df["Sixes"])
axes[2,0].set_title("Fours vs Sixes")
axes[2,0].set_xlabel("Fours")
axes[2,0].set_ylabel("Sixes")

# ------------------ 8. Balls vs Runs ------------------
axes[2,1].scatter(df["Balls_Faced"], df["Runs"])
axes[2,1].set_title("Balls vs Runs")
axes[2,1].set_xlabel("Balls Faced")
axes[2,1].set_ylabel("Runs")

# ------------------ 9. Result Count ------------------
df["Result"].value_counts().plot(
    kind='bar', ax=axes[2,2]
)
axes[2,2].set_title("Match Result Count")
axes[2,2].set_xlabel("Result")
axes[2,2].set_ylabel("Count")

# ------------------ Layout ------------------
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# group data
t20 = df[df["Match_Format"]=="T20"]["Runs"]
odi = df[df["Match_Format"]=="ODI"]["Runs"]
test = df[df["Match_Format"]=="Test"]["Runs"]

data = [t20, odi, test]

plt.figure(figsize=(10,6))

# ---------- Violin ----------
parts = plt.violinplot(data, showmeans=True, showmedians=True)

colors = ['skyblue','lightgreen','lightcoral']
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(colors[i])

# ---------- Box ----------
plt.boxplot(data, widths=0.15)

# labels
plt.xticks([1,2,3], ["T20","ODI","Test"])
plt.ylabel("Runs")
plt.title("Runs Distribution by Format (Violin + Box + Arrow)")

# ---------- Find Means ----------
means = [np.mean(t20), np.mean(odi), np.mean(test)]

# max & min
max_idx = np.argmax(means)
min_idx = np.argmin(means)

# ---------- Arrow (BEST PART ) ----------
plt.annotate(
    "Best Avg",
    xy=(max_idx+1, means[max_idx]),
    xytext=(max_idx+1.3, means[max_idx]+20),
    arrowprops=dict(facecolor='green', arrowstyle='->'),
    fontsize=10,
    color='green'
)

plt.annotate(
    "Low Avg",
    xy=(min_idx+1, means[min_idx]),
    xytext=(min_idx+1.3, means[min_idx]+20),
    arrowprops=dict(facecolor='red', arrowstyle='->'),
    fontsize=10,
    color='red'
)

plt.show()
```

```python
# Violin = pattern

# Box = stats

# Arrow = insight
```
