```python
import seaborn as sns

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

a = list("ABCDE")
b = np.array([10,20,30,40,50])
c = np.array([40,50,60,70,80])

plt.figure(figsize=(8,5))

# Base bars
ax = plt.barh(a, b, color=sns.color_palette("magma"), label="B")

# Stacked bars
ax1 = plt.barh(a, c, left=b, color=sns.color_palette("viridis"), label="C")

# Labels
plt.bar_label(ax, label_type='center')
plt.bar_label(ax1, label_type='center')

plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Horizontal Stacked Bar")

plt.legend()
plt.show()
```

| Graph Type          | Stacking Parameter |
| ------------------- | ------------------ |
| Vertical (`bar`)    | `bottom=`          |
| Horizontal (`barh`) | `left=`            |

```python
import matplotlib.pyplot as plt
import numpy as np

students = ["Aman", "Riya", "Karan", "Neha"]

maths = [80, 85, 78, 90]
science = [75, 88, 82, 85]

x = np.arange(len(students))   # positions
width = 0.35                   # bar width

plt.figure(figsize=(8,5))

# Bars
b1 = plt.bar(x - width/2, maths, width, label="Maths")
b2 = plt.bar(x + width/2, science, width, label="Science")

# Labels
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks Comparison")

plt.xticks(x, students)

# Value labels
plt.bar_label(b1, padding=3)
plt.bar_label(b2, padding=3)

plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()
```

# Histogram

```python
# Histogram ek graph hai jo continuous data ki frequency (kitni baar values aati hain) ko bins (ranges) me dikhata hai.
```

```python
# “Histogram = data ka distribution dikhata hai (values ka spread)”
```

```python
# 1. Continuous data ho

# Marks (0–100)

# Height (150–180 cm)

# Age

# 2. Distribution dekhna ho

# Data zyada kis range me hai

# data spread kaisa hai

#  3. Frequency samajhni ho

# Example:

# 0–10 → 2 students

# 10–20 → 5 students

# Example Samajh

# Marks:

# 10, 20, 25, 30, 35, 40, 45, 50

# Histogram:

# 0–20 → 2 values

# 20–40 → 4 values

# 40–60 → 2 values
```

| Term            | Meaning                    |
| --------------- | -------------------------- |
| Bins            | Value ranges (0–10, 10–20) |
| Frequency       | Count of values            |
| Continuous Data | Range wala data            |


```python
# Kab Use NA kare?

# Categories ho (A, B, C) → ❌ use bar graph

# exact comparison chahiye → ❌ bar better
```

| Feature | Histogram    | Bar Graph    |
| ------- | ------------ | ------------ |
| Data    | Continuous   | Categorical  |
| Gap     | No gap       | Gap hota hai |
| Use     | Distribution | Comparison   |

# Step Plot

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.step(x, y)

plt.show()
```

```python
# plt.step() = line ko stairs (steps) me draw karta hai
# normal plot() = smooth line
# step = zig-zag stairs
```

```python
plt.step(x, y, where='pre')   # left step
plt.step(x, y, where='post')  # right step
plt.step(x, y, where='mid')   # middle step
```

<!-- where	               Meaning
pre	step                pehle change hota hai
post	                baad me change hota hai
mid                  	beech me change -->

```python
# where	             Meaning
# pre	             step pehle change hota hai
# post	             baad me change hota hai
# mid	             beech me change
```

# . Styled Step Plot

```python
plt.step(x, y, where='mid', color='red', linestyle='--')

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Step Plot")

plt.grid(True)

plt.show()
```

# Step + Marker

```python
plt.step(x, y, where='mid', marker='o')
plt.show()
```

```python
# Kab use karte hain?

# Use step plot when:

# data suddenly change hota hai

# constant intervals ho

# ex:

# electricity usage

# stock price levels

# digital signals
```

```python
# Step plot = staircase style graph (sudden change dikhane ke liye)
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

fig, ax = plt.subplots(1,3, figsize=(10,4))

# Normal Line Plot
ax[0].plot(x, y, marker='o')
ax[0].set_title("Normal Plot")

# Step Plot
ax[1].step(x, y, where='pre')
ax[2].step(x, y, where='post')
ax[1].set_title("Step Plot")

plt.tight_layout()
plt.show()
```
