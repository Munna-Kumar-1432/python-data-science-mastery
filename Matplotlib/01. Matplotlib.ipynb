# plt.figure() creates a new figure (a blank canvas) where your plots will be drawn.

- Think of it as opening a new page before you start drawing a graph
- Create multiple separate plots
- Control the figure size
- Customize the overall plot container

```python
plt.figure(figsize=(4,3))
plt.plot(x,y)
plt.show()
```

```python
# labelling
# plt.plot(x, y)  x aur y values ka graph banata hai
# plt.xlabel('Class')  X-axis ka naam set karta hai
# plt.ylabel('Avg Marks')  Y-axis ka naam set karta hai
```

```python
plt.figure(figsize=(8,2))  ## 8 = length and 2 = width(height)
plt.plot(x,y)

plt.xlabel('Class')
plt.ylabel('Avg Marks')

plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.plot(x,y,alpha=0.3,label="Actual Marks")
plt.plot(x1,y1,alpha=0.9,linestyle='--',label="Estimated Marks")

plt.xlabel("Class",{"fontsize":15,"color":'r'})
plt.ylabel("Avg Marks",fontsize=20,color='g')

plt.title("My Scroe Card",fontsize=22,color='pink')
plt.grid(alpha=0.5)


for i,j in zip(x,y):
  plt.text(i,j-3,str(j))


plt.legend(loc=1)
plt.show()
```

```python
#loc  position of legend
# 0 = best
# 1 = upper right
# 2 = upper left
#3 = lower left
#4 = lower right
```

```python
plt.figure(figsize=(7,3))
plt.plot(x,y,alpha=0.3,label="Actual Marks")
plt.plot(x1,y1,alpha=0.9,linestyle='--',label="Estimated Marks",marker='o')

plt.xlabel("Class",{"fontsize":15,"color":'r'})
plt.ylabel("Avg Marks",fontsize=20,color='g')

plt.title("My Scroe Card",fontsize=22,color='pink')
plt.grid(alpha=0.5)


for i,j in zip(x,y):
  plt.text(i,j,str(j))

for i,j in zip(x1,y1):
    plt.text(i,j,str(j))


plt.legend(loc=4)
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.plot(x,y,alpha=0.3,label="Actual Marks",marker="o")
plt.plot(x1,y1,alpha=0.9,linestyle='--',label="Estimated Marks")

plt.xlabel("CLASS",{"fontsize":15,"color":'r'})
plt.ylabel("AVG MARKS",fontsize=20,color='g')

plt.title("SCORECARD",fontsize=22,color='pink')
plt.grid(alpha=0.5)


for i,j in zip(x,y):
  plt.text(i,j,str(j))

for i,j in zip(x1,y1):
    plt.text(i,j,str(j))


plt.xticks(color='r')
plt.yticks(color='g',fontsize=10)


plt.savefig("Class_Report1.png",dpi=1000)


plt.legend(loc=1)
plt.show()
```
