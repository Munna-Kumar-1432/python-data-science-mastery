```python
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120 Safari/537.36"
}

url = "https://www.ambitionbox.com/list-of-companies?page=2"

res = requests.get(url, headers=headers)
webpage = res.text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h2')[1]['title']
```

```python
companies = soup.find_all('h2', class_='companyCardWrapper__companyName')
```

```python
titles = [c.get('title') for c in companies]
```

```python
titles
```

```python
ratings = soup.find_all('div', class_='rating_text rating_text--md')
rating=[]
for r in ratings:
    rating.append(r.text.strip())
```

```python
rating
```

```python
interlinking = soup.find_all('span',class_='companyCardWrapper__interLinking')
```

```python
links= []
for i in interlinking:
    links.append(i.text.strip())
```

```python
links
```

```python
rating_values = soup.find_all('span',class_='companyCardWrapper__ratingValues')
```

```python
rat_val = []
for i in rating_values:
    rat_val.append(i.text.strip())
```

```python
rat_val
```

```python
company_logo = soup.find_all('div',class_='companyCardWrapper__companyLogo')
```

```python
image_urls = [c.find('img')['src'] for c in company_logo]

image_urls
```

```python
len(image_urls)
```

```python
data = []

for t, r, l,m in zip(titles, rating, links,image_urls):
    data.append({
        "company": t,
        "rating": r,
        "link": l,
        "image":m
    })

df = pd.DataFrame(data)
```
