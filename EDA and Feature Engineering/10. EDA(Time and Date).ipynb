# 1. Date-Time Creation & Conversion

| Operation         | Code                                      | Description                |
| ----------------- | ----------------------------------------- | -------------------------- |
| String → Datetime | `pd.to_datetime("2026-04-03")`            | Convert string to datetime |
| Column conversion | `df['date'] = pd.to_datetime(df['date'])` | Convert column             |
| Current date      | `pd.Timestamp.now()`                      | Get current time           |
| Custom date       | `pd.Timestamp('2025-01-01')`              | Create date                |


# 2. Extract Date Components (.dt)

| Feature     | Code                       | Output |
| ----------- | -------------------------- | ------ |
| Year        | `df['date'].dt.year`       | 2025   |
| Month       | `df['date'].dt.month`      | 1–12   |
| Day         | `df['date'].dt.day`        | 1–31   |
| Hour        | `df['date'].dt.hour`       | 0–23   |
| Minute      | `df['date'].dt.minute`     | 0–59   |
| Second      | `df['date'].dt.second`     | 0–59   |
| Day Name    | `df['date'].dt.day_name()` | Monday |
| Day of Week | `df['date'].dt.dayofweek`  | 0–6    |


# 3. Advanced Time Features

| Feature     | Code                               | Meaning    |
| ----------- | ---------------------------------- | ---------- |
| Weekend     | `df['date'].dt.dayofweek >= 5`     | Sat/Sun    |
| Quarter     | `df['date'].dt.quarter`            | 1–4        |
| Day of Year | `df['date'].dt.dayofyear`          | 1–365      |
| Week Number | `df['date'].dt.isocalendar().week` | Week index |
| Month Start | `df['date'].dt.is_month_start`     | True/False |
| Month End   | `df['date'].dt.is_month_end`       | True/False |


# 04. Date Difference

| Task              | Code                                                                         | Output       |
| ----------------- | ---------------------------------------------------------------------------- | ------------ |
| Days difference   | `(today - df['date']).dt.days`                                               | Integer days |
| Months difference | `(today.year - df['date'].dt.year)*12 + (today.month - df['date'].dt.month)` | Months       |
| Years difference  | `(today - df['date']).dt.days // 365`                                        | Years        |


# 05. Date Arithmetic

| Operation     | Code                                   |
| ------------- | -------------------------------------- |
| Add days      | `df['date'] + pd.Timedelta(days=7)`    |
| Subtract days | `df['date'] - pd.Timedelta(days=3)`    |
| Add months    | `df['date'] + pd.DateOffset(months=1)` |
| Add years     | `df['date'] + pd.DateOffset(years=1)`  |


# 6. Filtering Dates

| Task           | Code                               |
| -------------- | ---------------------------------- |
| After date     | `df[df['date'] > '2024-01-01']`    |
| Before date    | `df[df['date'] < '2025-01-01']`    |
| Specific month | `df[df['date'].dt.month == 1]`     |
| Weekend data   | `df[df['date'].dt.dayofweek >= 5]` |
