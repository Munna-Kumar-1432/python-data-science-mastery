# Bivariate Analysis

```python
# Analyzing the relationship between two variables
```

```python
# “Bi” = Two
# “Variate” = Variables
```

# Types of Bivariate Analysis (3 Main Types)

# a.Numerical vs Numerical

```python
# Examples: Age vs Salary, Height vs Weight

# Techniques / Graphs:

#         Scatter Plot
#         Line Plot
#         Correlation (Pearson)

#  Purpose:
#         To find relationship, trend, or correlation
```

# Categorical vs Numerical

```python
# One variable is category, one is number

# Examples: Gender vs Salary, City vs Income

# Techniques / Graphs:

#         Box Plot
#         Bar Chart
#         Violin Plot

# Purpose:
#   To compare distribution or averages across categories
```

# Categorical vs Categorical

```python
# Examples: Gender vs Department, City vs Product

# Techniques / Graphs:

#         Bar Chart (Grouped/Stacked)
#         Count Plot
#         Heatmap (Cross-tab)
#  Purpose:
#   To see association or relationship between categories
```

```python
# Bivariate = 2 variables
#     Used to find:
#     Relationship
#     Correlation
#     Patterns
```

| Type                       | Variables           | Example          | Graph        |
| -------------------------- | ------------------- | ---------------- | ------------ |
| Numerical vs Numerical     | Number + Number     | Age vs Salary    | Scatter Plot |
| Categorical vs Numerical   | Category + Number   | Gender vs Salary | Box Plot     |
| Categorical vs Categorical | Category + Category | Gender vs City   | Bar Chart    |

# 1. Scatterplot (Numerical - Numerical) 

```python
tips.head(2)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style=tips['smoker'],size=tips['size'])
plt.show()
```

# 2. Bar Plot (Numerical - Categorical)

```python

titanic.head()
```

```python
sns.barplot(data=titanic , x='Pclass' , y = 'Sex')
plt.show()
```

```python
sns.barplot(data=titanic , x='Pclass' , y = 'Sex' , hue='Sex')
plt.show()
```

```python
sns.barplot(data=titanic , x='Pclass' , y = 'Age' , hue='Sex')
plt.show()
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data=titanic, x='Sex', y='Age', hue='Survived')
plt.show()
```

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
plt.show()
```

# 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
sns.kdeplot(titanic[titanic['Survived']==0]['Age'], label='Not Survived')
sns.kdeplot(titanic[titanic['Survived']==1]['Age'], label='Survived')
```

```python
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True)
plt.show()
```

```python
sns.displot(data=titanic, x='Age', hue='Survived', kde=True)
plt.show()
```

```python
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True)
plt.show()
```

# 5. HeatMap (Categorical - Categorical)

```python

titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
plt.show()
```

# 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
plt.show()
```
