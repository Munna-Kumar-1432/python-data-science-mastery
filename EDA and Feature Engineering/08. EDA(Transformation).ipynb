```python
# Step 1: Dataset Load
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Select useful columns
df = df[['survived', 'pclass', 'sex', 'age', 'fare']]

# Handle missing values
df['age'].fillna(df['age'].mean(), inplace=True)

# Convert categorical → numeric
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
```

```python
df
```

```python
# Step 2: Visualize Raw Data
```

```python
plt.figure()
plt.hist(df['fare'], bins=30)
plt.title("Fare Distribution (Before Transformation)")
plt.show()
```

```python
# Fare data is skewed (right skewed)
# ML models confuse ho sakte hain
```

```python
# Step 3: Transformations Apply
```

```python
# . Log Transformation
```

```python
df['fare_log'] = np.log1p(df['fare'])

plt.figure()
plt.hist(df['fare_log'], bins=30)
plt.title("Fare After Log Transformation")
plt.show()
```

```python
# Large values compress karta hai
# Skew reduce karta hai
```

```python
# Min-Max Scaling
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['age_scaled', 'fare_scaled']] = scaler.fit_transform(df[['age', 'fare']])
```

```python
df
```

```python
# Standardization
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age_std', 'fare_std']] = scaler.fit_transform(df[['age', 'fare']])
```

```python
df
```

```python
# Polynomial Features
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(df[['age', 'fare']])
```

```python
poly_features
```

```python
# PCA (Dimensionality Reduction)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_features = pca.fit_transform(df[['age_scaled', 'fare_scaled']])
```

```python
pca_features
```

```python
# Step 4: Train Models
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df[['pclass', 'sex', 'age_scaled', 'fare_scaled']]
y = df['survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print("Logistic Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

y_pred_tree = tree.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_tree))
```

# Power Transformation

```python
# Power Transformation is a mathematical technique used to:

# Make data more normal (Gaussian-like)
# Reduce skewness
# Stabilize variance
```

```python
# Why Do We Need It?

# Real-world data is often:

# Skewed (not symmetric)
# Unevenly distributed
# Hard for models to interpret
```

# 1. Box-Cox Transformation

```python
# Works only for positive data (y > 0)
# Finds best λ (lambda) automatically
```

# Yeo-Johnson Transformation

```python
# Works for both positive and negative values
# More flexible than Box-Cox
```

```python
# How It Works (Intuition)

#          Suppose your data is right-skewed (long tail on right)

# Power transformation:

#         Compresses large values
#         Expands small values

#  Result: more balanced distribution
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer

# Sample skewed data
data = np.random.exponential(scale=2, size=1000)

# Apply Power Transformation
pt = PowerTransformer(method='yeo-johnson')
data_transformed = pt.fit_transform(data.reshape(-1, 1))

# Plot
plt.figure()
plt.hist(data, bins=30, alpha=0.5, label='Original')
plt.hist(data_transformed, bins=30, alpha=0.5, label='Transformed')
plt.legend()
plt.title("Power Transformation Effect")
plt.show()
```
