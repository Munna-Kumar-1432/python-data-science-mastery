```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E", "E"],
    "Age": [20, 21, 19, None, 22, 22],
    "Marks": [85, 90, 78, 88, None, None]
}

df = pd.DataFrame(data)
```

```python
df
```

| Step | Question            | Code                    | Example Output         | Explanation                |
| ---- | ------------------- | ----------------------- | ---------------------- | -------------------------- |
| 1️⃣  | Data size           | `df.shape`              | `(6, 3)`               | 6 rows, 3 columns          |
| 2️⃣  | Data preview        | `df.head()`             | A, B, C rows           | First few rows dekhte hain |
| 3️⃣  | Data types          | `df.dtypes`             | Name→object, Age→float | Column ka type pata chalta |
| 4️⃣  | Missing values      | `df.isnull().sum()`     | Age→1, Marks→2         | Kitne null values hain     |
| 5️⃣  | Statistical summary | `df.describe()`         | mean, std, min, max    | Numeric data ka summary    |
| 6️⃣  | Duplicate rows      | `df.duplicated().sum()` | `1`                    | Ek duplicate row hai (E)   |
| 7️⃣  | Correlation         | `df.corr()`             | Age vs Marks ≈ 0.5     | Relationship between cols  |
