# Univariate Analysis

```python
# Univariate = One Variable Analysis
```

```python
# It means:
#     You analyze only one column (feature) at a time
#     You don’t compare it with other columns
```

```python
# If you analyze only the Age column,
# that is Univariate Analysis
```

# What do we check in Univariate Analysis?

```python
# 01.Distribution
#         How values are spread
#         Example: Are most values high, low, or normal?
# 02. Central Tendency
#         Mean → average
#         Median → middle value
#         Mode → most frequent value
# 03. Spread (Variation)
#         Min / Max
#         Range
#         Standard deviation
# 04. Outliers
#         Unusual or extreme values
```
