```python
df1 = pd.DataFrame({
    "Name": ["Aman", "Rohit"],
    "Marks": [80, 85]
})

df2 = pd.DataFrame({
    "Name": ["Priya", "Neha"],
    "Marks": [90, 95]
})
```

```python
result = pd.concat([df1, df2])
display(result)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
result = pd.concat([df1, df2],ignore_index=True)
display(result)
```

```python
df1 = pd.DataFrame({
    "Name": ["Aman", "Rohit"],
    "Marks": [80, 85]
})

df2 = pd.DataFrame({
    "City": ["Delhi", "Mumbai"]
})

result = df1.join(df2)
display(result)
```
