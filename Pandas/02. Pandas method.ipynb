```python
# It is list of list
data = [
    ['Mohan', 234, 'Delhi'],
    ['Sahil', 345, 'Punjab'],
    ['Deepak', 444, 'Bihar']
]
```

```python
pd.DataFrame(data,columns=('Col_'+str(i) for i in range(len(data[0]))))
```

```python
pd.DataFrame(data,columns=('Col_'+str(i+4) for i in range(len(data[0]))))
```

```python
pd.DataFrame(data,index=(range(1,len(data)+1)),columns=('Col_'+str(i) for i in range(len(data[0]))))
```

```python
pd.DataFrame(data,index=range(10,len(data)+10),
             columns=['Col_'+str(i) for i in range(len(data[0]))])
```
