# How to find out missing percentage
- missing_percentage = df.isnull().sum() * 100 / len(df)
- df.isnull()  Checks for missing values (True/False)
- .sum()  Counts missing values in each column
- len(df)  Total number of rows
- X 100  Converts to percentage
