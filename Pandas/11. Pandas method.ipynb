```python
df1 = pd.DataFrame({
    "id": [1,2,3,4,5,6],
    "Science": [50,60,70,80,90,100],
    "Math": [55,65,75,85,95,105]
})
```

```python
df2 = pd.DataFrame({
    "id": [4,5,6,7,8],
    "Sports": [40,50,60,70,80],
    "Music": [45,55,65,75,85]
})
```

```python
df1.merge(df2, on="id", how="inner") 
# Inner Join (default)
# Common ids = 4,5,6
```

```python
df1.merge(df2, on="id", how="left")
# df1 ke sab ids rahenge (1–6)
```

```python
df1.merge(df2, on="id", how="right")
# df2 ke sab ids rahenge (4–8)
```

```python
df1.merge(df2, on="id", how="outer")
# Sab ids (1–8)
```

```python
# Join Type	          Matlab
# inner	             Sirf common
# left	             Left wale sab
# right	             Right wale sab
# outer	             Dono ke sab
```

```python
# Pivot kya hota hai?
# Pivot = Data ko rotate / reshape karna
# Rows ko columns bana dena
# Data ka structure change karna
```

```python
df = pd.DataFrame({
    "Student": ["A","A","B","B","C","C"],
    "Subject": ["Math","Science","Math","Science","Math","Science"],
    "Marks": [90,85,70,75,88,92]
})
```

```python
df
```

```python
# Pivot kya karega?

# Ab hum chahte hain:

# Student rows me ho

# Subject columns ban jaye

# Marks values me aa jaye
```

```python
df.pivot(index="Student", columns="Subject", values="Marks")
```

```python
df2 = pd.DataFrame({
    "Student": ["A","A","A","B","B"],
    "Subject": ["Math","Math","Science","Math","Science"],
    "Marks": [90,80,85,70,75]
})

df2
```

```python
df2.pivot_table(
    index="Student",
    columns="Subject",
    values="Marks",
    aggfunc="mean"
)
```

```python
# aggfunc="sum"
# aggfunc="count"
```

```python
# Pivot	                                Pivot Table
# Duplicate allowed nahi	           Duplicate allowed
# Sirf reshape karta	               Reshape + calculation karta
# Simple use	                       Advanced use
```

```python
# pivot() = Data ghumana
# pivot_table() = Data ghumana + calculation
```

```python
df = pd.DataFrame({
    "City": ["Delhi","Delhi","Delhi","Mumbai","Mumbai","Mumbai","Pune","Pune"],
    "Product": ["Phone","Laptop","Phone","Phone","Laptop","Laptop","Phone","Laptop"],
    "Sales": [100,200,150,120,250,300,130,220]
})

df
```

```python
df.pivot_table(
    index="City",
    columns="Product",
    values="Sales",
    aggfunc="sum"
)
```

```python
df.pivot_table(
    index="City",
    columns="Product",
    values="Sales",
    aggfunc="mean"
)
```

```python
df.pivot_table(
    index="City",
    columns="Product",
    values="Sales",
    aggfunc=["sum","mean"]
)
```

```python
df.pivot_table(
    index="City",
    columns="Product",
    values="Sales",
    aggfunc="sum",
    margins=True
)
```

```python
# pivot() = reshape only
# pivot_table() = reshape + aggregation + duplicates handle
```
