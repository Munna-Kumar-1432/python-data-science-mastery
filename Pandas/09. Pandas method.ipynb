# Timezone
- tz_localize() : Timezone attach karo
- tz_convert()  :Timezone change karo
- tz  timezone specify karne ke liye


```python
pd.date_range(start="2024-01-01", periods=3, freq="D", tz="Asia/Kolkata")
```

```python
pd.date_range(start="2024-01-01", periods=3, freq="H", tz="UTC")
```

```python
dates = pd.date_range("2024-01-01", periods=3)
dates.tz_localize("Asia/Kolkata")
```

```python
dates = pd.date_range("2024-01-01", periods=3, tz="UTC")
dates.tz_convert("Asia/Kolkata")
```
