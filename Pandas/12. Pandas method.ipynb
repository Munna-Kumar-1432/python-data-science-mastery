| Feature                | `map()`                               | `apply()`                       |
| ---------------------- | ------------------------------------- | ------------------------------- |
| Kaam kya karta hai     | Values ko replace/transform karta hai | Custom logic apply karta hai    |
| Kis par kaam karta hai | Sirf **Series (1 column)**            | Series + DataFrame (row/column) |
| Multiple columns use?  | Nahi                                |  Haan                          |
| Speed                  |  Fast                                | Thoda slow                   |
| Input                  | dict / function / Series              | function                        |
| Use case               | Mapping, replacing values             | Complex decision making         |


# pd.concat()

- Use: DataFrames ko vertically (upar-niche) ya horizontally (side-by-side) jodne ke liye

```python
df1 = pd.DataFrame({
    "player": ["Virat", "Rohit"],
    "runs": [120, 80]
})

df2 = pd.DataFrame({
    "player": ["Dhoni", "Gill"],
    "runs": [70, 60]
})
```

```python
df1
```

```python
df2
```

```python
result = pd.concat([df1,df2])
```

```python
result
```

# DataFrame.join()

-Use: Index ke basis par join karta hai

```python
df1 = pd.DataFrame({
    "player": ["Virat", "Rohit"]
})

df2 = pd.DataFrame({
    "runs": [120, 80]
})

result = df1.join(df2)
print(result)
```

```python
result
```

# DataFrame.merge()

-Use: SQL join jaisa (common column par)

```python
df1 = pd.DataFrame({
    "player": ["Virat", "Rohit"],
    "team": ["RCB", "MI"]
})

df2 = pd.DataFrame({
    "player": ["Virat", "Rohit"],
    "runs": [120, 80]
})

result = pd.merge(df1, df2, on="player")
```

```python
result
```

# DataFrame.apply()

- Use: Custom logic apply karna

```python
df = pd.DataFrame({
    "player": ["Virat", "Rohit"],
    "runs": [120, 40]
})

def performance(row):
    return "Good" if row["runs"] > 50 else "Bad"

df["performance"] = df.apply(performance, axis=1)
```

```python
df
```

# DataFrame.insert()

- Use: Specific position par new column add karna

```python
df = pd.DataFrame({
    "player": ["Virat", "Rohit"],
    "runs": [120, 80]
})

df.insert(1, "team", ["RCB", "MI"])
```

```python
df
```

| Function   | Purpose (Kya karta hai)                                              | Syntax                            | Works On         | Key Parameters           | Example Use Case                | Output Type      | Important Notes                         |
| ---------- | -------------------------------------------------------------------- | --------------------------------- | ---------------- | ------------------------ | ------------------------------- | ---------------- | --------------------------------------- |
| `concat()` | Multiple DataFrames ko **stack/join** karta hai (row ya column wise) | `pd.concat([df1, df2], axis=0/1)` | DataFrame        | `axis`, `ignore_index`   | Multiple files ko combine karna | DataFrame        | axis=0 → rows add, axis=1 → columns add |
| `join()`   | **Index ke basis par** do DataFrames join karta hai                  | `df1.join(df2)`                   | DataFrame        | `how`, `on`              | Jab index same ho               | DataFrame        | Default = left join                     |
| `merge()`  | SQL jaisa **common column par join**                                 | `pd.merge(df1, df2, on="col")`    | DataFrame        | `on`, `how`              | Database-style join             | DataFrame        | inner, left, right, outer joins support |
| `apply()`  | Custom function ko **row/column par apply** karta hai                | `df.apply(func, axis=0/1)`        | Series/DataFrame | `axis`                   | Condition / logic lagana        | Series/DataFrame | axis=1 → row-wise, axis=0 → column-wise |
| `insert()` | Specific position par **new column add** karta hai                   | `df.insert(loc, col, values)`     | DataFrame        | `loc`, `column`, `value` | Column ko beech me add karna    | None (in-place)  | Existing column name hua toh error      |
