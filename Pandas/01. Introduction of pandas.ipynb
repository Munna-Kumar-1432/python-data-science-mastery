# Dropna and Fillna
- Both are used to handle missing values (NaN) in Pandas.
- dropna() : Removes the missing values (NaN).
- fillna() : dropna() deletes rows where value is missing.

```python
s = pd.Series([10, 20, np.nan, 40, np.nan, 60])
print(s)
```

```python
s.dropna()
```

```python
s.dropna().reset_index(drop=True) ## Drop
```

```python
s.fillna(0) ## Fill by zero
```

```python
s.fillna(s.mean()) ## Replaces NaN with average value.
```

```python
s
```

```python
s.fillna(method="ffill") ## Takes previous value.
```

```python
s.fillna(method="bfill") ## Takes next value.
```

```python
s.fillna(4)
```
