# Step : 06 Model Evaluation

```python

# 1. Regression Problem
#    - MSE
#    - MAE
#    - RMSE
#    - R² Score
#    - Adjusted R²
#    - p-value
#    - F-statistic

# 2. Classification Problem(Binary , Multiclass)
#    - Accuracy
#    - Precision
#    - Recall
#    - F1 Score
#    - Sensitivity
#    - Specificity
#    - Confusion Matrix
#    - Classification Report
#    - Type 1 Error
#    - Type 2 Error 
```

```python
# Binary class Confucion matrix
# ML model ---> Predict patient Cancer yes or no

# Actual Scenario vs Model Prediction

# 1 Patient actual : Cancer Yes , Model : Cancer Yes  ---> [No mistake]  (True Positive)

# 2 Patient actual : Cancer No , Model : Cancer No  ---> [No mistake]  (True Negative)

# 3 Patient actual : Cancer No , Model : Cancer Yes  ---> [Mistake]  (False Positive / Type 1 Error)

# 4 Patient actual : Cancer Yes , Model : Cancer No  ---> [Mistake]  (False Negative / Type 2 Error)

# id ML model score tie : we will select ML which has Type 1 Error : lowest
```

```python
# TP → Sick ko Sick bola
# TN → Healthy ko Healthy bola
# FP → Healthy ko Sick bola
# FN → Sick ko Healthy bola
```

# Accuracy

- Accuracy batata hai ki model ne total predictions me se kitni correct predictions ki.

```python
# Accuracy = (TP + TN) / (TP + TN + FP + FN)

# TP → True Positive

# TN → True Negative

# FP → False Positive

# FN → False Negative
```
