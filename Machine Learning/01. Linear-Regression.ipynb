```python
 # Linear Regression

# 1 ya zyada input features se
# Continuous output predict karta hai

# Example:

# Study hours → Marks predict karna

# Experience → Salary predict karna

# Ad spend → Sales predict karna
```

```python
# Linear Regression Kab Use Karte Hain?


# ⃣ Target variable numeric ho (continuous)
# ⃣ Relationship roughly linear ho
# ⃣ Scatter plot straight-line pattern dikhaye
```

```python
# Kab Use Nahi Karte?

# Agar target Yes/No ho → Logistic Regression

# Agar relation curve jaisa ho → Polynomial Regression

# Agar classification problem ho
```

```python
# Linear Regression Kab Confirm Karein?

# Checklist:

# Target numeric hai?
# Scatter plot straight line jaisa hai?
# Correlation high hai?
```

```python
# I used Linear Regression because the target variable was continuous and the scatter plot showed a strong 
# linear relationship between the independent and dependent variables.
```

```python
# Equation Kaise Banti Hai? (y = mx + c)

# Linear Regression ka basic formula hota hai:
#  y = mx + c

# Machine Learning language me:
#  y=b1x+b0

# y → predicted value (Marks)

# x → input feature (Study hours)

# m / b1 → slope (coefficient)

# c / b0 → intercept


# Coefficient Ka Meaning

# Coefficient = slope = change in y when x increases by 1 unit.
```

# Residual = Actual−Predicted

```python
# Least Squares Method = ∑(Actual−Predicted)2
```

```python
# Linear Regression fits a straight line that minimizes squared residuals. T
#     he coefficient represents the change in the target variable for a one-unit change in the predictor. 
#     Residuals are the differences between actual and predicted values.
```

```python
# Error Loss or Cost Function
```

```python
# Three type od linear regreesion error
# MAE(Mean Absolute Errro)
# MSE(Mean Square Error)
# RMSE(Root mean square error)
```

```python
# MAE=n1​∑∣Actual−Predicted∣  (1/n = n1)

# MSE=n1​∑(Actual−Predicted)2 (1/n = n1)

# RMSE=(MSE)1/2
```

```python
# Metric	          Outlier Effect	         Unit	           Interpretation
# MAE	                  Low	                 Same as target	     Average error
# MSE	                  High	                 Squared unit	     Large error punish
# RMSE	                  High	                 Same as target	     More practical
```

```python
MAE = (compare_df['Actual_y:'] - compare_df['Model Predicated y:']).abs().mean()
```

```python
MAE
```

```python
MSE = ((compare_df['Actual_y:'] - compare_df['Model Predicated y:'])**2).mean()
```

```python
MSE
```

```python
import math
RMSE = math.sqrt(MSE)
```
