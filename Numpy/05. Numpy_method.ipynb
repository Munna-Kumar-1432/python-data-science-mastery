```python
# vstack()  Vertical Stack (add to top to bottom)
# Condition: Columns same should be same
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])
```

```python
np.vstack((a,b))
```

```python
# . hstack()  Horizontal Stack (side by side add)
# Condition: Rows should be same
```

```python
result = np.hstack((a, b))
print(result)
```

```python
# dstack()  Depth Stack (Add in 3 dimensional)
```

```python
a,b
```

```python
result = np.dstack((a, b))
print(result)
```

```python
# | Function | Join Direction | Axis   | Result Example Shape |
# | -------- | -------------- | ------ | -------------------- |
# | vstack   | top-bottom    | axis 0 | (4,2)                |
# | hstack   | Left-Right     | axis 1 | (2,4)                |
# | dstack   | Depth (layer)  | axis 2 | (2,2,2)              |
```

```python
np.where(a > 25) ## which index value is true
```

```python
b = np.array([[1, 5],
              [8, 2]])

np.where(b > 4) # (array([rows]), array([columns]))
```

```python
# | Dimension | np.where() return karta hai |
# | --------- | --------------------------- |
# | 1D        | 1 array (indices)           |
# | 2D        | 2 arrays (row, col)         |
# | 3D        | 3 arrays (depth, row, col)  |
# | nD        | n arrays                    |
```
