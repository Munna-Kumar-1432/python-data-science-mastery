# Perceptron
- A Perceptron is one of the simplest machine learning algorithms.
- It is a binary classifier, meaning it predicts only two classes:
- A perceptron is a model that takes inputs, applies weights, and makes a decision.
- y=w1​x1​+w2​x2​+⋯+b
  

```python
# y=w1​x1​+w2​x2​+⋯+b

# x = input features

# w = weights

# b = bias

# If output > 0 → class 1
# Else → class 0
```

```python
# Perceptron is the simplest form of a neural network used for binary classification with a linear decision boundary
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
```

```python
p
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
# w1 = 40.26 , w2 = -36 : Apply on Perceptron formula
```

```python
p.intercept_
```

```python
# b(bias) =  -25 : Apply on perceptron formula
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

plot_decision_regions(X.values, y.values, clf=p,legend=2)
plt.show()
```
