# if-elif-else statement

```python
num = 7

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
#   if condition1:
#     statement1
#   elif condition2:
#     statement2
#   elif condition3:
#     statement3
#   else:
#     statement4
```

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

| Statement      | Use                   |
| -------------- | --------------------- |
| `if`           | Check one condition   |
| `if-else`      | Two possible outcomes |
| `if-elif-else` | Multiple conditions   |


```python
# Nested if Statement

# A nested if means an if statement inside another if statement.
```

```python
# Ternary Operator (Conditional Expression)

# A ternary operator is a short form of if-else.
```

```python
age = 18

result = "Adult" if age >= 18 else "Minor"

print(result)
```

| Feature     | Nested if                               | Ternary Operator           |
| ----------- | --------------------------------------- | -------------------------- |
| Purpose     | Multiple conditions inside another `if` | Short form of `if-else`    |
| Code length | Longer                                  | Short                      |
| Readability | Good for complex logic                  | Good for simple conditions |


```python
# List of students allowed for GATE exam
students = ["Rahul", "Aman", "Ravi", "Munna", "Suresh"]

# Take input from user
name = input("Enter your name: ").title()

# Check if name is in the list
if name in students:
    print("You are allowed to sit in the GATE exam")
else:
    print("You are NOT allowed to sit in the GATE exam")
```

# Python program to check whether a letter is a Vowel or Consonant

```python
letter = input("Enter a letter: ").lower()

if letter in ['a', 'e', 'i', 'o', 'u']:
    print("It is a vowel")
else:
    print("It is a consonant")
```

```python
ch = input("Enter a letter: ").lower()

if ch in "aeiou":
    print("Vowel")
else:
    print("Consonant")
```

# Using if with membership operators (in, not in)

```python
names = ["Rahul", "Aman", "Munna", "Ravi"]

name = input("Enter your name: ").title()

if name not in names:
    print("You are not registered")
```

| Operator | Meaning                        | Example                 |
| -------- | ------------------------------ | ----------------------- |
| `in`     | Checks if value exists         | `"Munna" in names`      |
| `not in` | Checks if value does not exist | `"Rakesh" not in names` |


# Logical operators in conditions (and, or, not)

```python
# if condition1 and condition2:
#     statement
```

```python
age = 20
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")
```

```python
# if condition1 or condition2:
#     statement
```

```python
day = "Sunday"

if day == "Saturday" or day == "Sunday":
    print("Weekend")
```

```python
# Reverses the condition (True becomes False, False becomes True).
```

```python
# if not condition:
#     statement
```

| Condition1 | Condition2 | `and` | `or`  |
| ---------- | ---------- | ----- | ----- |
| True       | True       | True  | True  |
| True       | False      | False | True  |
| False      | True       | False | True  |
| False      | False      | False | False |


```python
age = 16
has_id = True

if age >= 18 and has_id:
    print("Allowed")
elif age < 18 or not has_id:
    print("Not allowed")
```

#  Using if with identity operators (is, is not)

```python
# . is Operator

# Returns True if both variables refer to the same object.
```

```python
# if a is b:
#     statement
```

```python
a = 10
b = 10

if a is b:
    print("Both variables refer to the same object")
```

```python
# is not Operator

# Returns True if variables refer to different objects.
```

```python
# if a is not b:
#     statement
```

```python
a = [1,2,3]
b = [1,2,3]

if a is not b:
    print("Different objects in memory")
```

```python
data = None

if data is None:
    print("No data available")
```

| Operator | Meaning                                      |
| -------- | -------------------------------------------- |
| `==`     | Checks if **values are equal**               |
| `is`     | Checks if **objects are the same in memory** |


# Pass statement in conditional blocks

```python
# pass Statement in Conditional Blocks (Python)


# The pass statement is a null statement.
# It means do nothing, but it allows the program to run without errors.

# It is often used when a statement is required syntactically, but you don’t want to write code yet.
```

```python
# if condition:
#     pass
```

```python
num = 10

if num > 5:
    pass
else:
    print("Number is small")
```

```python
num = 0

if num > 5:
    pass
else:
    print("Number is small")
```

```python
# If num > 5, Python executes pass → nothing happens.

# If the condition is false → else block runs.
```

```python
name = input("Enter your name: ")

if name == "Munna":
    pass
else:
    print("Welcome", name)
    
```

| Use                | Explanation                               |
| ------------------ | ----------------------------------------- |
| Placeholder        | Code will be written later                |
| Avoid syntax error | Python requires a statement inside blocks |
| Empty block        | Keeps program structure valid             |


| Keyword    | Meaning                    | Effect in Loop          | Example             |
| ---------- | -------------------------- | ----------------------- | ------------------- |
| `pass`     | Does nothing (placeholder) | Loop continues normally | `if x==5: pass`     |
| `continue` | Skips current iteration    | Moves to next iteration | `if x==5: continue` |
| `break`    | Stops the loop completely  | Exits the loop          | `if x==5: break`    |


```python
for i in range(1,6):
    if i == 3:
        pass
    print(i)
```

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

```python
for i in range(1,6):
    if i == 3:
        break
    print(i)
```

| Keyword    | Remember   |
| ---------- | ---------- |
| `pass`     | Do nothing |
| `continue` | Skip       |
| `break`    | Stop loop  |
