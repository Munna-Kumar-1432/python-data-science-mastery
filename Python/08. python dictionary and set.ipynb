```python
# remove()

# Removes an element.
# If the element is not found, it gives an error.
```

```python
s = {1, 2, 3}
s.remove(2)

print(s)
```

```python
# discard()

# Removes an element.
# If the element is not present, no error occurs.
```

```python
s = {1, 2, 3}
s.discard(4)
```

```python
# pop()

# Removes and returns a random element from the set.
```

```python
s = {1, 2, 3}
s.pop()

print(s)
```

```python
# clear()

# Removes all elements from the set.
```

```python
s = {1, 2, 3}
s.clear()

print(s)
```

```python
# union()

# Combines elements from both sets.
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
```

```python
# intersection()

# Returns common elements from both sets.
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```

```python
# difference()

# Returns elements present in the first set but not in the second.
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.difference(b))
```

```python
# symmetric_difference()

# Returns elements that are in either set but not in both.
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.symmetric_difference(b))
```

```python
# issubset()

# Checks if all elements of one set exist in another.
```

```python
a = {1, 2}
b = {1, 2, 3, 4}

print(a.issubset(b))
```

```python
a = {1, 2,5}
b = {1, 2, 3, 4}

print(a.issubset(b))
```

```python
# issuperset()

# Checks if a set contains all elements of another set.
```

```python
b = {1, 2, 3, 4}
a = {1, 2}

print(b.issuperset(a))
```

```python
# isdisjoint()

# Returns True if two sets have no common elements.
```

```python
a = {1, 2}
b = {3, 4}

print(a.isdisjoint(b))
```

```python
# copy()

# Creates a copy of a set.
```

```python
a = {1, 2, 3}

b = a.copy()

print(b)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)  # union
print(a & b)  # intersection
print(a - b)  # difference
print(a ^ b)  # symmetric difference
```

```python
list_example = [1,2,3,3]
tuple_example = (1,2,3,3)
set_example = {1,2,3,3}

print(list_example)
print(tuple_example)
print(set_example)
```

```python
# When to use what

# Use List → when data changes frequently

# Use Tuple → when data should not change

# Use Set → when you need unique values
```

```python
# intersection_update() updates the original set with the common elements of another set.

# It keeps only the elements that exist in both sets.
```

```python
a = {1,2,3,4}
b = {3,4,5,6}

a.intersection_update(b)

print(a)
```

```python
# intersection()	            Returns a new set
# intersection_update()	       Modifies the original set
```

```python
a = {1,2,3}
b = {2,3,4}

c = a.intersection(b)

print(c)
print(a)
```

```python
# update() adds elements from another set (or iterable) into the original set.
```

```python
a = {1,2,3}
b = {3,4,5}

a.update(b)

print(a)
```

```python
# difference_update() removes elements from the first set that are present in the second set.
```

```python
a = {1,2,3,4}
b = {3,4,5}

a.difference_update(b)

print(a)
```

```python
# symmetric_difference_update() keeps elements that are not common in both sets.

# It removes common elements and keeps only unique elements from both sets.
```

```python
a = {1,2,3}
b = {3,4,5}

a.symmetric_difference_update(b)

print(a)
```

```python
 
# update()                               =	Adds elements from another set
# difference_update()                    =	Removes common elements from first set
# symmetric_difference_update()	         =     Keeps only non-common elements
```

| Method                   | Description                                  | Example                     |
| ------------------------ | -------------------------------------------- | --------------------------- |
| `add()`                  | Adds one element to the set                  | `s.add(5)`                  |
| `update()`               | Adds multiple elements from another iterable | `s.update([4,5])`           |
| `remove()`               | Removes an element (error if not found)      | `s.remove(2)`               |
| `discard()`              | Removes an element (no error if not found)   | `s.discard(2)`              |
| `pop()`                  | Removes and returns a random element         | `s.pop()`                   |
| `clear()`                | Removes all elements                         | `s.clear()`                 |
| `copy()`                 | Returns a copy of the set                    | `new = s.copy()`            |
| `union()`                | Returns all elements from both sets          | `a.union(b)`                |
| `intersection()`         | Returns common elements                      | `a.intersection(b)`         |
| `difference()`           | Elements in first set but not second         | `a.difference(b)`           |
| `symmetric_difference()` | Elements not common in both sets             | `a.symmetric_difference(b)` |
| `issubset()`             | Checks if a set is a subset of another       | `a.issubset(b)`             |
| `issuperset()`           | Checks if a set contains another set         | `a.issuperset(b)`           |


```python
a = {1,2,3}
b = {3,4,5}

print(a.union(b))                 # {1,2,3,4,5}
print(a.intersection(b))          # {3}
print(a.difference(b))            # {1,2}
print(a.symmetric_difference(b))  # {1,2,4,5}
```

| Method                   | Description                           | Example Code                                  | Result            |
| ------------------------ | ------------------------------------- | --------------------------------------------- | ----------------- |
| `add()`                  | Adds one element                      | `s={1,2}; s.add(3)`                           | `{1,2,3}`         |
| `update()`               | Adds multiple elements                | `s={1,2}; s.update([3,4])`                    | `{1,2,3,4}`       |
| `remove()`               | Removes element (error if not found)  | `s={1,2,3}; s.remove(2)`                      | `{1,3}`           |
| `discard()`              | Removes element (no error if missing) | `s={1,2,3}; s.discard(4)`                     | `{1,2,3}`         |
| `pop()`                  | Removes random element                | `s={1,2,3}; s.pop()`                          | `{2,3}` (example) |
| `clear()`                | Removes all elements                  | `s={1,2}; s.clear()`                          | `set()`           |
| `copy()`                 | Creates a copy of set                 | `a={1,2}; b=a.copy()`                         | `{1,2}`           |
| `union()`                | Combines two sets                     | `a={1,2}; b={2,3}; a.union(b)`                | `{1,2,3}`         |
| `intersection()`         | Common elements                       | `a={1,2}; b={2,3}; a.intersection(b)`         | `{2}`             |
| `difference()`           | Elements in first not second          | `a={1,2,3}; b={2}; a.difference(b)`           | `{1,3}`           |
| `symmetric_difference()` | Elements not common                   | `a={1,2}; b={2,3}; a.symmetric_difference(b)` | `{1,3}`           |
| `issubset()`             | Checks if subset                      | `a={1,2}; b={1,2,3}; a.issubset(b)`           | `True`            |
| `issuperset()`           | Checks if superset                    | `a={1,2,3}; b={1,2}; a.issuperset(b)`         | `True`            |
| `isdisjoint()`           | No common elements                    | `a={1,2}; b={3,4}; a.isdisjoint(b)`           | `True`            |


| Method                          | Description               | Example                                              | Result  |
| ------------------------------- | ------------------------- | ---------------------------------------------------- | ------- |
| `intersection_update()`         | Keeps common elements     | `a={1,2,3}; b={2,3,4}; a.intersection_update(b)`     | `{2,3}` |
| `difference_update()`           | Removes common elements   | `a={1,2,3}; b={2}; a.difference_update(b)`           | `{1,3}` |
| `symmetric_difference_update()` | Keeps non-common elements | `a={1,2}; b={2,3}; a.symmetric_difference_update(b)` | `{1,3}` |


# Python Dictionary Methods Table

| Method         | Description                                 | Example Code                     | Result                          |
| -------------- | ------------------------------------------- | -------------------------------- | ------------------------------- |
| `clear()`      | Removes all items                           | `d={'a':1,'b':2}; d.clear()`     | `{}`                            |
| `copy()`       | Returns a copy of dictionary                | `d={'a':1}; x=d.copy()`          | `{'a':1}`                       |
| `fromkeys()`   | Creates dictionary with given keys          | `dict.fromkeys(['a','b'],0)`     | `{'a':0,'b':0}`                 |
| `get()`        | Returns value of key                        | `d={'a':1}; d.get('a')`          | `1`                             |
| `items()`      | Returns key-value pairs                     | `d={'a':1,'b':2}; d.items()`     | `dict_items([('a',1),('b',2)])` |
| `keys()`       | Returns all keys                            | `d={'a':1,'b':2}; d.keys()`      | `dict_keys(['a','b'])`          |
| `values()`     | Returns all values                          | `d={'a':1,'b':2}; d.values()`    | `dict_values([1,2])`            |
| `pop()`        | Removes element by key                      | `d={'a':1,'b':2}; d.pop('a')`    | `{'b':2}`                       |
| `popitem()`    | Removes last inserted pair                  | `d={'a':1,'b':2}; d.popitem()`   | `('b',2)`                       |
| `setdefault()` | Returns value if key exists, else inserts   | `d={'a':1}; d.setdefault('b',2)` | `{'a':1,'b':2}`                 |
| `update()`     | Updates dictionary with new key-value pairs | `d={'a':1}; d.update({'b':2})`   | `{'a':1,'b':2}`                 |


```python
d = {'name':'Rahul', 'age':21}

print(d.keys())
print(d.values())
print(d.items())

d.update({'city':'Delhi'})
print(d)

d.pop('age')
print(d)
```
