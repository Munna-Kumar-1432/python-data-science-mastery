# Class vs Object

| Feature | Class                         | Object                          |
| ------- | ----------------------------- | ------------------------------- |
| Meaning | Blueprint / template          | Real instance of class          |
| Purpose | Defines properties & methods  | Uses those properties & methods |
| Memory  | No memory allocated initially | Memory is allocated             |
| Example | `Student`                     | `s1 = Student()`                |
