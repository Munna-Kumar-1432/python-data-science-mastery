#  One linear if else

```python
# value_if_true if condition else value_if_false
```

```python
age = 20
result = "Adult" if age >= 18 else "Minor"
print(result)
```

```python
num = 7
print("Even") if num % 2 == 0 else print("Odd")
```

```python
marks = 40
status = "Pass" if marks >= 35 else "Fail"
print(status)
```

```python
print("Munna is allowed") if age>=18 else print("Munna not allowed")
```

# Number Guessing Game

```python
import random
secret_number = random.randint(1,10)

guess = int(input("Guess the number (1-10): "))

if guess == secret_number:
    print("🎉 Bilkul sahi jawab!")
    print("💰 Aap jeet gaye... 7 CRORE RUPAYE!!!")
    print("🔥 Computer ji, inko 7 crore rupaye diye jaayein!")
else:
    print("❌ Yeh jawab galat hai...")
    print(f"Computer ye ne choice kiye tha {secret_number}")
    print("💔 Afsos! Aap 7 crore se chook gaye.")
    print("🔁 Kripya phir se koshish kijiye.")
```

# Login Credential Check

```python
print("🔐 Login System")


correct_username = "admin"
correct_password = "1234"


username = input("Enter Username: ")
password = input("Enter Password: ")


if username == correct_username:
    if password == correct_password:
        print("✅ Login Successful! Welcome", username)
    else:
        print("❌ Invalid Password")
else:
    print("❌ Invalid Username")
```

# Rock, Paper, or Scissors

```python
import random

print("🎮 Rock Paper Scissors Game")

choices = ['rock', 'paper', 'scissors']

user = input("Enter rock, paper, or scissors: ").lower()

computer = random.choice(choices)

print("Computer chose:", computer)

if user == computer:
    print("🤝 It's a draw!")
elif user == "rock" and computer == "scissors":
    print("🎉 You win!")
elif user == "paper" and computer == "rock":
    print("🎉 You win!")
elif user == "scissors" and computer == "paper":
    print("🎉 You win!")
else:
    print("💻 Computer wins!")
```
