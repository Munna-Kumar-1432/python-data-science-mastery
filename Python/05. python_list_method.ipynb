# Lists are Mutable (Can be changed)

```python
fruits = ["apple", "banana", "orange"]

# Change an element
fruits[1] = "grape"
print(fruits)  # Output: ['apple', 'grape', 'orange']

# Add elements
fruits.append("mango")
print(fruits)  # Output: ['apple', 'grape', 'orange', 'mango']

# Remove elements
fruits.remove("apple")
print(fruits)  # Output: ['grape', 'orange', 'mango']
```

# Adding Elements 

```python
my_list = [1, 2, 3]

# append() - adds at the end
my_list.append(4)
print(my_list)  # [1, 2, 3, 4]

# insert() - adds at specific position
my_list.insert(1, 10)
print(my_list)  # [1, 10, 2, 3, 4]

# extend() - adds multiple elements
my_list.extend([5, 6, 7])
print(my_list)  # [1, 10, 2, 3, 4, 5, 6, 7]
```

```python

numbers = [5, 2, 8, 1, 9, 3, 2]

# Sorting
numbers.sort()
print(numbers)  # [1, 2, 2, 3, 5, 8, 9]

# Reverse
numbers.reverse()
print(numbers)  # [9, 8, 5, 3, 2, 2, 1]

# Count occurrences
print(numbers.count(2))  # 2

# Find index
print(numbers.index(5))  # 2 (position of 5)

# Length of list
print(len(numbers))  # 7
```

```python


original = [1, 2, 3]

# Shallow copy (be careful!)
copy1 = original  # This just creates a reference
copy1[0] = 100
print(original)  # [100, 2, 3] - Original also changed!

# Correct ways to copy
copy2 = original.copy()      # Method 1
copy3 = list(original)       # Method 2  
copy4 = original[:]          # Method 3

copy2[0] = 200
print(original)  # [100, 2, 3] - Original unchanged
print(copy2)     # [200, 2, 3]
```

```python

numbers = ['a', 'b', 'c', 'd', 'e', 'f']

# Index:    0    1    2    3    4    5

print(numbers[:4:-1])  # ['f']        (index 5)

print(numbers[:3:-1])  # ['f', 'e']   (indices 5,4)


print(numbers[:2:-1])  # ['f', 'e', 'd'] (indices 5,4,3)

print(numbers[:1:-1])  # ['f', 'e', 'd', 'c'] (indices 5,4,3,2)

print(numbers[:0:-1])  # ['f', 'e', 'd', 'c', 'b'] (indices 5,4,3,2,1)
```
