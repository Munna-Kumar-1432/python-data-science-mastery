# Palindrom

```python
word  =input("Enter is Anything ....")

# word = str(word)

print("Palindrome") if word.lower() == word.lower()[::-1] else print("Not Palindrome")
```

```python
data = input("Enter is Anything...")
ans = ""

for i in range(-1,-len(data)-1,-1):
    ans += data[i]
if ans == data:
    print("PALINDROME")
else:
    print("NOT PALINDROME")
```

# Fibonacci Series

```python
# The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones, typically starting with 0 and 1. 
```

```python
n = input("Up to what number should u generate fibonacci series")

n = int(n)

fib = [0,1]

for i in range(n-2):
    next_num = fib[-1] + fib[-2]
    fib.append(next_num)
print(fib)
```

```python
n = int(input("Up to what number should genrate fibonacci series"))
a,b = 0,1
for i in range(n):
    print(a,end=" ")
    a,b = b,a+b
```

```python
def fibonacci(limit):
    a, b = 0, 1
    
    while a <= limit:
        print(a, end=" ")
        a, b = b, a + b


fibonacci(50)
```

# Anagram

```python
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)
```

```python
print(is_anagram("listen", "silent"))
```

```python
print(is_anagram("hello", "world"))
```

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    
    count = {}
    
    for char in s1:
        count[char] = count.get(char, 0) + 1
        
    for char in s2:
        if char not in count or count[char] == 0:
            return False
        count[char] -= 1
        
    return True
```

```python
is_anagram("Munna","aunnM")
```

#  Prime Number

```python
def is_prime(n):
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


print(is_prime(7))  
print(is_prime(10))  
```

```python
import math

def is_prime(n):
    if n <= 1:
        return False
    
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(10)
```

```python
def prime_range(n):
    for num in range(2, n+1):
        if is_prime(num):
            print(num, end=" ")

prime_range(20)
```

# Armstrong Number

```python
def is_armstrong(n):
    s = str(n)
    power = len(s)
    
    total = 0
    for digit in s:
        total += int(digit) ** power
        
    return total == n


print(is_armstrong(153))  
print(is_armstrong(123))   
```

```python
def is_armstrong(n):
    temp = n
    digits = 0
    
    while temp > 0:
        digits += 1
        temp //= 10
    
    temp = n
    total = 0
    
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10
        
    return total == n
    
```

```python
is_armstrong(41)
```

```python
def armstrong_range(n):
    for num in range(1, n+1):
        if is_armstrong(num):
            print(num, end=" ")

armstrong_range(1000)
```
