#  Function Scope in python

```python
# Scope means the region where a variable is accessible
```

# Types of Scope in Python (LEGB Rule)

 | Scope | Meaning   |
| ----- | --------- |
| L     | Local     |
| E     | Enclosing |
| G     | Global    |
| B     | Built-in  |


# 1. Local Scope

 - Variables defined inside a function

```python
def my_func():
    x = 10  # local variable
    print(x)

my_func()
```

```python
# x is only accessible inside the function
```

# 2. Global Scope

- Variables defined outside all functions

```python
x = 20  # global

def show():
    print(x)

show()
```

```python
# Accessible everywhere
```

# Modify Global Variable

```python
x = 10

def change():
    global x
    x = 50

change()
print(x)  
```

# Enclosing Scope (Closure Scope)

- Variables in outer function, used by inner function

```python
def outer():
    x = 100

    def inner():
        print(x)

    inner()

outer()
```

```python
# inner() can access x
```

# Modify Enclosing Variable

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()  
```

# 4. Built-in Scope

- Predefined functions in Python

```python
print(len("hello"))
```

# LEGB Order (Very Important)
When Python searches for a variable:

- 01 Local

- 02 Enclosing

- 03 Global

- 04 Built-in

```python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print(x)

    inner()

outer()
```

```python
x = "Global Player"

def outer():
    x = "Munna (Enclosing)"

    def inner():
        x = "Match Player (Local)"
        print(x)

    inner()
    print(x)

outer()
print(x)
```

```python
# Scope defines where a variable can be used
```

```python
# Python uses LEGB rule to resolve variable names
```

```python
player = "Global Munna"   # Global Scope

def team():
    player = "Team Munna"   # Enclosing Scope

    def match():
        player = "Match Munna"   # Local Scope
        print("Inside match:", player)

    match()
    print("Inside team:", player)

team()
print("Outside:", player)
```

```python
player = "Global Munna"

def team():
    player = "Team Munna"

    def match():
        nonlocal player
        player = "Updated Munna"

    match()
    print("Inside team:", player)

team()
print("Outside:", player)
```

```python
player = "Global Munna"

def team():
    player = "Team Munna"

    def match():
        nonlocal player
        player = "Updated Munna"

    match()
    print("Inside team:", player)

team()
print("Outside:", player)
```

```python
# nonlocal → changes enclosing variable

# global → changes global variable
```

```python
# “Local wins first, then Enclosing, then Global”
```

| Scope             | Where Defined         | Accessible Where          | Keyword to Modify | Example                  | Priority   |
| ----------------- | --------------------- | ------------------------- | ----------------- | ------------------------ | ---------- |
| **Local (L)**     | Inside a function     | Only inside that function | ❌ Not needed      | `x = 10` inside function | 🥇 Highest |
| **Enclosing (E)** | Inside outer function | Inner (nested) functions  | `nonlocal`        | outer → inner            | 🥈         |
| **Global (G)**    | Outside all functions | Entire program            | `global`          | `x = 20`                 | 🥉         |
| **Built-in (B)**  | Python predefined     | Everywhere                | ❌ Not needed      | `len(), sum()`           | 🏁 Lowest  |


| Scope     | Variable Value   | Where Used            | Output         |
| --------- | ---------------- | --------------------- | -------------- |
| Local     | `"Match Munna"`  | Inside inner function | Printed first  |
| Enclosing | `"Team Munna"`   | Inside outer function | Printed second |
| Global    | `"Global Munna"` | Outside function      | Printed last   |


| Keyword    | Used For                  | Scope Affected | Example      |
| ---------- | ------------------------- | -------------- | ------------ |
| `global`   | Modify global variable    | Global         | `global x`   |
| `nonlocal` | Modify enclosing variable | Enclosing      | `nonlocal x` |
| None       | Local variable            | Local          | `x = 10`     |


```python
# Python uses LEGB (Local → Enclosing → Global → Built-in) to find variables
```

```python
player = "Global Munna"

def season():
    player = "Season Munna"
    total_runs = 0

    def match(runs):
        nonlocal total_runs
        player = "Match Munna"

        total_runs += runs
        print(f"{player} scored {runs} runs")

    match(50)
    match(30)

    print("Total Runs in Season:", total_runs)
    print("Player in season:", player)

season()
print("Player globally:", player)
```

# Generator

```python
# A generator function is a function that:

# Uses the keyword yield instead of return

# Returns values one by one (lazy evaluation)

# Does not store all values in memory
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# How it Works

# yield pauses the function

# Saves its state

# Next time it resumes from where it stopped
```

# Generator vs Normal Function

| Feature   | Normal Function      | Generator Function  |
| --------- | -------------------- | ------------------- |
| Keyword   | `return`             | `yield`             |
| Returns   | Whole result at once | One value at a time |
| Memory    | High                 | Low                 |
| Execution | Runs completely      | Pauses & resumes    |


```python
def munna_runs():
    runs = [10, 50, 30]

    for r in runs:
        yield r

gen = munna_runs()

for run in gen:
    print("Munna scored:", run)
```

```python
def infinite_numbers():
    n = 1
    while True:
        yield n
        n += 1

gen = infinite_numbers()

for i in range(5):
    print(next(gen))
```

```python
# Why Use Generators?

# Saves memory

# Handles large data

# Used in data streaming, file reading, APIs
```

```python
# Important Points

# yield makes function a generator

# You can use next() to get next value

# Once finished → raises StopIteration
```

```python
# Generator = function that gives values one by one using yield
```

```python
def my_generator(limit):
    for i in range(1, limit+1):
        yield i

for i in my_generator(3):
    print(i)
```

```python
# Generator remembers state

# yield = pause + return

# Next call resumes from same line
```

```python
# First iteration

# i = 1

# yield 1 → value return

# function pause

# print : 1

# Second iteration

# function resumes

# i = 2

# yield 2 → return + pause

#  print: 2

# Third iteration

# resume again

# i = 3

# yield 3

# print: 3

# Important Clarification

#  yield loop ko return nahi karta
#  yield caller (for loop / next()) ko value return karta hai

#  Real Meaning

#  yield =
# ye value le lo, main yahin ruk raha hoon, baad me yahin se start karunga”



#  yield = return + pause + resume from same point
```

| Feature          | `return`                              | `yield`                                 |
| ---------------- | ------------------------------------- | --------------------------------------- |
| Purpose          | Ends the function and returns a value | Returns a value and pauses the function |
| Execution        | Runs the whole function at once       | Runs step-by-step                       |
| Memory Usage     | Higher (stores all data)              | Lower (lazy evaluation)                 |
| State            | Does not remember state               | Remembers state                         |
| Function Type    | Normal function                       | Generator function                      |
| Output           | Direct value (list, int, etc.)        | Generator object                        |
| Resume Execution | ❌ Cannot resume                       | ✅ Resumes from last point               |
| Iteration        | Not automatic                         | Works easily with loops                 |


```python
def normal():
    return [1, 2, 3]

print(normal())
```

```python
def generator():
    yield 1
    yield 2
    yield 3

for i in generator():
    print(i)
```

| Step | `return`        | `yield`                  |
| ---- | --------------- | ------------------------ |
| 1    | Function starts | Function starts          |
| 2    | Runs completely | Runs until first `yield` |
| 3    | Returns value   | Returns value and pauses |
| 4    | Function ends   | Resumes later            |
| 5    | No state saved  | State is preserved       |


```python
def munna_runs():
    yield 10
    yield 20
    yield 30

for r in munna_runs():
    print("Run:", r)
```

# Recursion

```python
# Recursion is when a function calls itself to solve a problem.

# Instead of solving a big problem at once, we:
# Break it into smaller sub-problems
# Solve each using the same function
```

```python
# Every recursive function MUST have:

# 1. Base Case (Stopping condition)

# This stops recursion
# Without it → infinite loop ❌

# 2. Recursive Case

# Function calls itself with smaller input
```

```python
def factorial(n):
    if n == 1:   # Base case
        return 1
    return n * factorial(n - 1)  # Recursive call

print(factorial(5))
```

```python
# # How it Works (Call Stack)

# factorial(5)
# = 5 * factorial(4)
# = 5 * 4 * factorial(3)
# = 5 * 4 * 3 * factorial(2)
# = 5 * 4 * 3 * 2 * factorial(1)
# = 5 * 4 * 3 * 2 * 1
# = 120
```

```python
# Recursive Formula:
# fib(n) = fib(n-1) + fib(n-2)
```

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)

print(fib(6))
```

```python
# Call Stack Concept

# When recursion happens:

# Each function call is stored in memory (stack)
# Last call finishes first (LIFO)
```

```python
# f(3)
#  → f(2)
#    → f(1)
#      → return
```

```python
# When to Use Recursion?

# Use it when:

# Problem can be divided into smaller similar problems
# Tree / Graph problems
# Backtracking (like Sudoku, N-Queens)
# Divide & Conquer (Merge Sort, Quick Sort)
```

| Feature | Recursion     | Loop   |
| ------- | ------------- | ------ |
| Code    | Clean & short | Longer |
| Speed   | Slower        | Faster |
| Memory  | More (stack)  | Less   |


| Pattern          | Key Idea      |
| ---------------- | ------------- |
| Linear           | One call      |
| Binary           | Two calls     |
| Multiple         | Many choices  |
| Backtracking     | Try + Undo    |
| Divide & Conquer | Split + Merge |

```python
def subseq(arr, index, path):
    if index == len(arr):
        print(path)
        return
    
    subseq(arr, index+1, path+[arr[index]])  # include
    subseq(arr, index+1, path)               # exclude
```
