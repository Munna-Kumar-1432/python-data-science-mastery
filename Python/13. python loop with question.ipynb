# Infinite Loop

```python
# while True:
#     print("Hello")
```

| Feature                   | `for` Loop                                                          | `while` Loop                                                |
| ------------------------- | ------------------------------------------------------------------- | ----------------------------------------------------------- |
| **Definition**            | Used to iterate over a sequence (list, tuple, string, range, etc.). | Used to repeat a block of code while a condition is `True`. |
| **Condition Handling**    | The number of iterations is usually known beforehand.               | The loop runs until a specified condition becomes `False`.  |
| **Syntax**                | `for variable in sequence:`                                         | `while condition:`                                          |
| **Control Variable**      | Automatically handled by the loop.                                  | Must be manually initialized and updated.                   |
| **Use Case**              | Best when you know how many times the loop should run.              | Best when the number of iterations is unknown.              |
| **Example**               | `python\nfor i in range(5):\n    print(i)\n`                        | `python\ni = 0\nwhile i < 5:\n    print(i)\n    i += 1\n`   |
| **Risk of Infinite Loop** | Rare, because iteration is controlled.                              | Higher risk if the condition never becomes `False`.         |


```python
# for loop → when the number of iterations is known.
# while loop → when the condition controls the loop
```

# Nested Loop

```python
row = 1
while row <= 5:         
    col = 1
    while col <= row:    
        print('*', end='')
        col += 1
    print()             
    row += 1
```

```python
for i in range(1,11):
    for j in range(1,i):
        print("*",end="")
    print()
    
        
```

| Keyword      | Purpose                     | What it does in a loop                                                                                    | Example Code                                                                     | Output      |
| ------------ | --------------------------- | --------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------- | ----------- |
| **break**    | Stops the loop immediately  | When the condition is met, the loop terminates and control moves outside the loop                         | `python\nfor i in range(1,6):\n    if i == 3:\n        break\n    print(i)\n`    | `1 2`       |
| **continue** | Skips the current iteration | When the condition is met, the rest of that iteration is skipped and the loop moves to the next iteration | `python\nfor i in range(1,6):\n    if i == 3:\n        continue\n    print(i)\n` | `1 2 4 5`   |
| **pass**     | Placeholder statement       | Does nothing; used when a statement is required syntactically but you don’t want any action               | `python\nfor i in range(1,6):\n    if i == 3:\n        pass\n    print(i)\n`     | `1 2 3 4 5` |


```python
# break → completely stops the loop.

# continue → skips the current iteration and goes to the next one.

# pass → does nothing, it is just a placeholder.
```

```python
for i in range(5):
    if i == 2:
        continue
    if i == 4:
        break
    print(i)
```

# Check if a Number is Prime

```python
num = 7

for i in range(2, num):
    if num % i == 0:
        print("Not Prime")
        break
else:
     print("Prime")
```

```python
num = 200
prime = []
import math

for i in range(2,num + 1):
    for j in range(2,int(math.sqrt(i)+1)):
        if i % j == 0:
            break;
    else:
        prime.append(i)
print("Total Number Of Prime Number  : ",prime)
print("Total Number of Prime Number : ",len(prime))
```

```python
import math

num = 200
prime = []

if num <= 1:
    print(f"{num} is not a Prime Number")
else:

    for i in range(2, num + 1):
        for j in range(2, int(math.sqrt(i)) + 1):
            if i % j == 0:
                break
        else:
            prime.append(i)

print("Total Prime Numbers :", prime)
print("Count of Prime Numbers :", len(prime))
```

```python
import math

num = 100
prime = []

if num <= 1:
    print(f"{num} is not a Prime Number")
else:

    i = 2
    while i <= num:

        j = 2
        while j <= int(math.sqrt(i)):
            if i % j == 0:
                break
            j += 1
        else:
            prime.append(i)

        i += 1

print("Total Prime Numbers :", prime)
print("Count of Prime Numbers :", len(prime))
```

# Find the Largest Number in a List

```python
arr = [3, 7, 2, 9, 5]

max_num = arr[0]

for num in arr:
    if num > max_num:
        max_num = num

print(max_num)
```

# Find the Smallest Number in a List

```python
arr = [8, 4, 6, 2, 10]

min_num = arr[0]

for num in arr:
    if num < min_num:
        min_num = num

print(min_num)
```

# Sum of All Elements in a List

```python
arr = [1,2,3,4]

total = 0

for num in arr:
    total += num

print(total)
```

# Count Even and Odd Numbers

```python
arr = [1,2,3,4,5,6]

even = 0
odd = 0

for num in arr:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even =", even)
print("Odd =", odd)
```

# Reverse a List Using Loop

```python
arr = [1,2,3,4]

rev = []

for i in range(len(arr)-1, -1, -1):
    rev.append(arr[i])

print(rev)
```

# Find Duplicate Elements

```python
arr = [1,2,3,2,4,5,1]

dup = []

for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[i] == arr[j] and arr[i] not in dup:
            dup.append(arr[i])

print(dup)
```

# Remove Duplicates from List

```python
arr = [1,2,2,3,4,4,5]

unique = []

for num in arr:
    if num not in unique:
        unique.append(num)

print(unique)
```

# Find Second Largest Number

```python
arr = [10,20,4,45,99]

first = second = -1

for num in arr:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num

print(second)
```
