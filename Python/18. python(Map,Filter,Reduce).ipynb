```python
import random

def kbc_game():
    questions = [
        {
            "question": "What is the capital of India?",
            "options": ["A. Mumbai", "B. Delhi", "C. Kolkata", "D. Chennai"],
            "answer": "B"
        },
        {
            "question": "Who is known as the Father of Computers?",
            "options": ["A. Charles Babbage", "B. Alan Turing", "C. Elon Musk", "D. Bill Gates"],
            "answer": "A"
        },
        {
            "question": "Which language is used for web apps?",
            "options": ["A. Python", "B. Java", "C. JavaScript", "D. C++"],
            "answer": "C"
   Syntax     },
        {
            "question": "Which planet is known as the Red Planet?",
            "options": ["A. Earth", "B. Mars", "C. Jupiter", "D. Saturn"],
            "answer": "B"
        }
    ]

    prize_money = [1000, 5000, 10000, 50000]

    lifelines = {
        "50-50": True,
        "audience": True,
        "phone": True
    }

    print("Welcome to KBC Game")

    for i, q in enumerate(questions):
        print(f"\nQuestion {i+1} for Rs {prize_money[i]}")
        print(q["question"])

        for opt in q["options"]:
            print(opt)

        while True:
            print("\nAvailable Lifelines:")
            for key, val in lifelines.items():
                if val:
                    print(f"- {key}")

            choice = input("\nEnter option (A/B/C/D) or type 'lifeline': ").strip().upper()

            if choice == "LIFELINE":
                print("\nChoose Lifeline:")
                for key, val in lifelines.items():
                    if val:
                        print(f"- {key}")

                life = input("Enter lifeline name: ").strip().lower()

                if life == "50-50" and lifelines["50-50"]:
                    lifelines["50-50"] = False
                    correct = q["answer"]
                    options = ["A", "B", "C", "D"]
                    options.remove(correct)
                    remove = random.sample(options, 2)

                    print("\n50-50 Lifeline Used")
                    for opt in q["options"]:
                        if opt[0] not in remove:
                            print(opt)

                elif life == "audience" and lifelines["audience"]:
                    lifelines["audience"] = False
                    print("\nAudience Poll Results:")
                    for opt in ["A", "B", "C", "D"]:
                        percent = random.randint(10, 70)
                        print(f"{opt}: {percent}%")

                elif life == "phone" and lifelines["phone"]:
                    lifelines["phone"] = False
                    print("\nCalling your friend...")
                    if random.random() > 0.3:
                        print(f"Friend thinks answer is {q['answer']}")
                    else:
                        print("Friend is not sure")

                else:
                    print("Invalid or already used lifeline")

                continue

            elif choice in ["A", "B", "C", "D"]:
                if choice == q["answer"]:
                    print("Correct Answer")
                    break
                else:
                    print(f"Wrong Answer! Correct answer was {q['answer']}")
                    print(f"You won Rs {prize_money[i-1] if i > 0 else 0}")
                    return
            else:
                print("Invalid input! Try again.")

    print("\nCongratulations! You won the game!")


kbc_game()
```

# Filter

```python
# filter() is a built-in Python function

# It is used to select elements from an iterable based on a condition

# It returns only those elements for which the condition is True
```

```python
# filter(function, iterable)
# function → condition (returns True/False)

# iterable → list, tuple, etc.
# Returns a filter object, so convert it to list
```

```python
nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)

print(list(result))
```

```python
def is_even(x):
    return x % 2 == 0

nums = [1, 2, 3, 4]

print(list(filter(is_even, nums)))
```

```python
names = ["rahul", "aman", "sita", "raj"]

result = list(filter(lambda x: len(x) > 4, names))

print(result)
```

```python
data = [0, 1, "", "hello", None, True]

print(list(filter(None, data)))
```

```python
marks = [35, 80, 90, 20]

passed = list(filter(lambda x: x >= 40, marks))
```

# Reduce

```python
# reduce() is used to apply a function cumulatively to the items of an iterable
# It reduces the whole list into a single value
```

```python
# reduce() is not built-in directly
# You need to import it:
```

```python
from functools import reduce
```

```python
# reduce(function, iterable)
# function → takes 2 arguments
# iterable → list, tuple, etc.
```

```python
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)

print(result)
```

```python
# Step 1: 1 + 2 = 3  
# Step 2: 3 + 3 = 6  
# Step 3: 6 + 4 = 10 
```

```python
from functools import reduce

nums = [10, 20, 30]

print(reduce(lambda x, y: x + y, nums))
```

```python
from functools import reduce

nums = [1, 2, 3, 4]

print(reduce(lambda x, y: x * y, nums))
```

```python
from functools import reduce

nums = [5, 8, 2, 10]

max_val = reduce(lambda x, y: x if x > y else y, nums)

print(max_val)
```

```python
# reduce() Used?
#  1. Aggregation (Most Important)

#  Sum, product, max, min

#  2. Data Processing Pipelines

#  When combining results step by step

#  3. Functional Programming

#  Used with map() and filter()
```

```python
# reduce() applies a function cumulatively to elements of an iterable and reduces it to a single value
```

| Feature           | map()             | filter()          | reduce()                    | for loop        |
| ----------------- | ----------------- | ----------------- | --------------------------- | --------------- |
| Purpose           | Transform data    | Select data       | Combine data into one value | General purpose |
| Output            | Modified iterable | Filtered iterable | Single value                | Any result      |
| Function required | Yes               | Yes (condition)   | Yes (2 args)                | No              |
| Built-in          | Yes               | Yes               | No (`functools`)            | Yes             |
| Readability       | Medium            | Medium            | Hard                        | Easy            |
| Flexibility       | Limited           | Limited           | Limited                     | Very High       |
| Performance       | Fast              | Fast              | Fast                        | Slightly slower |
| Memory            | Efficient         | Efficient         | Efficient                   | Less efficient  |


| Function | Meaning                          |
| -------- | -------------------------------- |
| map()    | Apply function to each element   |
| filter() | Keep elements based on condition |
| reduce() | Combine all elements into one    |
| loop     | Do anything step by step         |


| Situation            | Best Choice |
| -------------------- | ----------- |
| Apply same operation | map()       |
| Filter data          | filter()    |
| Get single result    | reduce()    |
| Complex logic        | for loop    |


# Decorator

```python
# A decorator is a function that modifies or extends another function’s behavior
#  Without changing the original function code
```

```python
#A decorator is a function that takes another function and adds extra functionality to it
```

```python
# Why Use Decorators?
# Add logging 
# Add authentication 
# Measure execution time 
# Avoid repeating code
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper
```

```python
def say_hello():
    print("Hello")

say_hello = my_decorator(say_hello)

say_hello()
```

# Using @ Syntax (Shortcut)

```python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
```

```python
def my_decorator(func):
    def wrapper(name):
        print("Before")
        func(name)
        print("After")
    return wrapper

@my_decorator
def greet(name):
    print("Hello", name)

greet("Rahul")
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result
    return wrapper
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        print("Time:", end - start)
    return wrapper

@timer
def work():
    for i in range(1000000):
        pass

work()
```
