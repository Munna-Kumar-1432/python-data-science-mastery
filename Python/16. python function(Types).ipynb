```python
# Armstrong
```

```python
def is_armstrong(num):
    # convert number to string to count digits
    digits = len(str(num))
    
    sum = 0
    temp = num
    
    while temp > 0:
        digit = temp % 10
        sum += digit ** digits
        temp //= 10
    
    if sum == num:
        return True
    else:
        return False


# Example usage
number = 153

if is_armstrong(number):
    print(number, "is an Armstrong number")
else:
    print(number, "is not an Armstrong number")
```

```python
def is_armstrong(num):
    digits = len(str(num))
    total = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == num


# Range
start = 100
end = 500

print("Armstrong numbers between", start, "and", end, ":")

for i in range(start, end + 1):
    if is_armstrong(i):
        print(i)
```

# Fibonacci Series?

```python
def fibonacci(n):
    a, b = 0, 1

    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

# Example
fibonacci(10)
```

# Fibonacci using Recursion

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# Print series
for i in range(10):
    print(fib(i), end=" ")
```

```python
n = int(input("Enter number of terms: "))
fibonacci(n)
```

# Args 

```python
# *args is used to pass multiple arguments to a function
```

```python
# args stands for arguments

# * means it will take multiple values and store them as a tuple
```

```python
def show(*args):
    print(args)

show(1, 2, 3, 4)
```

```python
def print_numbers(*args):
    for num in args:
        print(num)

print_numbers(10, 20, 30)
```

```python
def add(*args):
    total = 0
    for num in args:
        total += num
    return total

print(add(1, 2, 3, 4))
```

```python
# *args allows variable number of arguments

# Values are stored as a tuple

# You can use any name instead of args (but args is standard)
```

```python
def func(a, *args):
    print(a)
    print(args)

func(1, 2, 3, 4)
```

#  Kwargs

```python
# **kwargs allows a function to accept any number of keyword arguments
# It stores them as a dictionary

# kwargs = keyword arguments

# ** → collects key=value pairs
```

```python
def show(**kwargs):
    print(kwargs)

show(name="Rahul", age=21)
```

```python
def show(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

show(name="Aman", age=22, city="Delhi")
```

```python
def student(**kwargs):
    print("Student Details:")
    for key, value in kwargs.items():
        print(key, "=", value)

student(name="Ravi", age=20, course="BTech")
```

```python
def func(a, b, **kwargs):
    print("a:", a)
    print("b:", b)
    print("Other:", kwargs)

func(1, 2, x=10, y=20)
```

```python
def create_user(**details):
    if "name" in details:
        print("Name:", details["name"])
    if "email" in details:
        print("Email:", details["email"])

create_user(name="John", email="john@gmail.com")
```

| Feature | *args   | **kwargs       |
| ------- | ------- | -------------- |
| Type    | Tuple   | Dictionary     |
| Input   | Values  | key=value      |
| Example | (1,2,3) | {"a":1, "b":2} |


| Feature               | `*args`                              | `**kwargs`                        |
| --------------------- | ------------------------------------ | --------------------------------- |
| Full Form             | Arguments                            | Keyword Arguments                 |
| Purpose               | Accept multiple positional arguments | Accept multiple keyword arguments |
| Data Type             | Tuple                                | Dictionary                        |
| Syntax                | `*args`                              | `**kwargs`                        |
| How values are passed | Without key (just values)            | With key=value pairs              |
| Example Call          | `func(1, 2, 3)`                      | `func(a=1, b=2)`                  |
| Inside Function       | `(1, 2, 3)`                          | `{'a': 1, 'b': 2}`                |
| Access Method         | Loop through tuple                   | Use `.items()` for key-value      |
| Flexibility           | Handles unknown number of values     | Handles unknown named arguments   |
| Use Case              | Math operations, lists of values     | Configs, APIs, user data          |
| Order in Function     | Comes before `**kwargs`              | Always last                       |
| Can rename?           | Yes (`*nums`)                        | Yes (`**data`)                    |

```python
def order_summary(customer_name, *items, **details):
    print("Customer:", customer_name)
    
    print("\nItems Ordered:")
    for item in items:
        print("-", item)
    
    print("\nOther Details:")
    for key, value in details.items():
        print(f"{key}: {value}")


# Functionnn callll
order_summary(
    "Munna",
    "Pizza", "Burger", "Coke",
    address="Delhi",
    payment="Online",
    delivery_time="30 mins"
)
```

```python
def match_summary(player_name, *runs, **details):
    print("Player:", player_name)
    
    print("\nRuns in each match:")
    for i, run in enumerate(runs, start=1):
        print(f"Match {i}: {run} runs")
    
    print("\nMatch Details:")
    for key, value in details.items():
        print(f"{key}: {value}")
    
    print("\nTotal Runs:", sum(runs))


# Function Call
match_summary(
    "Munna",
    45, 60, 30, 80,
    team="Game Changer",
    opponent="Punjab 11",
    venue="Champion Ground",
    format="T20"
)
```

```python
def batting_analysis(player, *scores, **info):
    print("Player:", player)
    
    print("\nScores:")
    for s in scores:
        print(s)
    
    print("\nStats:")
    print("Total Runs:", sum(scores))
    print("Average:", sum(scores)/len(scores))
    
    print("\nExtra Info:")
    for k, v in info.items():
        print(k, ":", v)


batting_analysis(
    "Munna",
    50, 70, 10, 90,
    team="India",
    format="ODI"
)
```

```python
def bowling_stats(player, *wickets, **details):
    print("Player:", player)
    
    print("\nWickets per match:", wickets)
    print("Total Wickets:", sum(wickets))
    
    print("\nDetails:")
    for k, v in details.items():
        print(k, ":", v)


bowling_stats(
    "Munna",
    2, 3, 1, 4,
    opponent="England",
    venue="Delhi"
)
```

```python
import pandas as pd

def create_match_df(player, *runs, **details):
    # Create match numbers
    matches = list(range(1, len(runs) + 1))
    
    # Create DataFrame
    df = pd.DataFrame({
        "Match": matches,
        "Player": player,
        "Runs": runs
    })
    
    # Add extra details as columns
    for key, value in details.items():
        df[key] = value
    
    return df


# Function Call
df = create_match_df(
    "Munna",
    45, 60, 30, 80,
    team="India",
    opponent="Australia",
    format="T20"
)

print(df)
```

# Nested Function

```python
# A nested function is a function defined inside another function
```

```python
# def outer():
#     print("Outer function")

#     def inner():
#         print("Inner function")

#     inner()  # calling inner function

# outer()
```

```python
def match_analysis(player, *runs):
    print("Player:", player)

    def calculate_stats():
        total = sum(runs)
        avg = total / len(runs)
        print("Total Runs:", total)
        print("Average:", avg)

    calculate_stats()


match_analysis("Munna", 40, 60, 80)
```

```python
# Why Use Nested Functions?

# 1. Code Organization

# - Helper functions inside main function

# 2. Data Hiding (Encapsulation)

# - Inner function is not accessible outside

# 3. Avoid Repetition

# - Use helper logic inside
```

```python
def outer():
    def inner():
        return "Hello from inner"
    return inner

func = outer()
print(func())
```

```python
def outer(x):
    def inner(y):
        return x + y
    return inner

add = outer(10)
print(add(5))
```
